```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import cv2
```

```python
imagepath='images.jpg'
```

read image with imread method

```python
img=cv2.imread(imagepath)
```

```python
img.shape # it's 3d arrray
```

Now we will convert the image in grayscale image bcs opencv use  oppisite layout BGR

```python
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

Now For face dectection we will use haarcascade method so we will load that classifier

```python
face_classifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

Now we will detect the image

```python
face=face_classifier.detectMultiScale(gray_image,scaleFactor=1.1,minNeighbors=5,minSize=(40,40))
```

Now image is already detected so i need to creat a box around face  

```python
for(x,y,w,h) in face:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

the model succesfully detecting face and creating a boundry box around face
